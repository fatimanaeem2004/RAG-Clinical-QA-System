# RAG Data Preparation and Segmenting
This notebook details the loading, processing, and chunking experiments performed on our domain-specific medical Guidelines corpus.
Specifically, we study how **Chunk Size** and **Chunk Overlap** affect context segmentation.

```python
# Verify imports
import os
import sys
sys.path.append('..')
from src.document_processor import DocumentProcessor
print("Processor library loaded successfully!")
```

### Load and Inspect the Corpus Documents
We read the guidelines text files from the `../data/` directory and print their character lengths.

```python
processor = DocumentProcessor()
data_dir = "../data"
docs = processor.load_documents(data_dir)
for name, text in docs.items():
    print(f"Document: {name} | Length: {len(text)} chars | Word Count: {len(text.split())} words")
```

### Chunking Strategy Experimentation
Here we test split dimensions: 
1. **Small Chunk** (150 tokens, 15 tokens overlap)
2. **Medium Chunk** (300 tokens, 30 tokens overlap)
3. **Large Chunk** (600 tokens, 60 tokens overlap)

```python
for size, overlap in [(150, 15), (300, 30), (600, 60)]:
    chunks = processor.process_directory(data_dir, chunk_size=size, chunk_overlap=overlap)
    print(f"[Experiment] Size: {size:3d} | Overlap: {overlap:2d} | Generated Chunks: {len(chunks):2d}")
    if chunks:
        avg_len = sum(c['token_count'] for c in chunks) / len(chunks)
        print(f"             -> Average actual chunk size: {avg_len:.1f} tokens")
```

### Inspecting Chunk Contents
Let's print the first chunk of a document to inspect text alignment and segmentation completeness.

```python
chunks = processor.process_directory(data_dir, chunk_size=300, chunk_overlap=30)
first_chunk = chunks[0]
print(f"Chunk ID: {first_chunk['chunk_id']}")
print(f"Token Count: {first_chunk['token_count']}")
print("Content Snippet:\n----------------")
print(first_chunk['text'][:350] + "...")
```
